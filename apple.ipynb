```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv('apple_products.csv')
```

```python
print(data.isnull().sum())
```

```python
data.describe()
```

```python
# will get the top 10 iphone which were sold out 

highest_rated = data.sort_values(by=["Star Rating"], ascending=False)

highest_rated = highest_rated.head(10)

print(highest_rated['Product Name'])
```

```python


# Count the occurrences of each iPhone model in the highest_rated dataset
iphone_counts = highest_rated["Product Name"].value_counts()

# Extract the iPhone model names as labels
labels = iphone_counts.index

# Get the number of ratings for each iPhone model
counts = highest_rated["Number Of Ratings"]

# Create a bar plot showing the number of ratings for each iPhone model
plt.figure(figsize=(10, 5))
plt.bar(labels, counts)
plt.title("Number of Ratings for Highest Rated iPhones")
plt.xlabel("iPhone Model")
plt.ylabel("Number of Ratings")
plt.xticks(rotation=45)
plt.show()
```

```python

# Count the occurrences of each iPhone model in the highest_rated dataset
iphone_counts = highest_rated["Product Name"].value_counts()

# Extract the iPhone model names as labels
labels = iphone_counts.index

# Get the number of reviews for each iPhone model
counts = highest_rated["Number Of Reviews"]

# Create a bar plot showing the number of reviews for each iPhone model
plt.figure(figsize=(10, 6))
plt.bar(labels, counts)
plt.title("Number of Reviews of Highest Rated iPhones")
plt.xlabel("iPhone Model")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot using matplotlib.pyplot
plt.figure(figsize=(10, 6))
plt.scatter(data["Number Of Ratings"], data["Sale Price"], s=data["Discount Percentage"], alpha=0.7)
plt.title("Relationship between Sale Price and Number of Ratings of iPhones")
plt.xlabel("Number Of Ratings")
plt.ylabel("Sale Price")
plt.grid(True)

# Add trendline using numpy polyfit
import numpy as np
m, b = np.polyfit(data["Number Of Ratings"], data["Sale Price"], 1)
plt.plot(data["Number Of Ratings"], m * data["Number Of Ratings"] + b, color='red')

# Add colorbar
plt.colorbar(label="Discount Percentage")

plt.show()
```

```python
# Scatter plot using matplotlib.pyplot
plt.figure(figsize=(10, 6))
plt.scatter(data["Number Of Ratings"], data["Discount Percentage"], s=data["Sale Price"], alpha=0.02)
plt.title("Relationship between Discount Percentage and Number of Ratings of iPhones")
plt.xlabel("Number Of Ratings")
plt.ylabel("Discount Percentage")
plt.grid(True)

# Add trendline using numpy polyfit
import numpy as np
m, b = np.polyfit(data["Number Of Ratings"], data["Discount Percentage"], 1)
plt.plot(data["Number Of Ratings"], m * data["Number Of Ratings"] + b, color='red')

# Add colorbar
plt.colorbar(label="Sale Price")

plt.show()
```
